```python
date_range = pd.date_range(start='2021-01-01', end='2021-12-31', freq='D')
cycles = 3  # Number of cycles in the year
cyclic_pattern = 150 + 100 * np.sin(np.linspace(0, cycles * 2 * np.pi, len(date_range)))

df_cyclic = pd.DataFrame({
    'Date': date_range,
    'Sales': cyclic_pattern
})

plt.figure(figsize=(12, 6))
plt.plot(df_cyclic['Date'], df_cyclic['Sales'], label='Cyclic Sales')
plt.title('Sales with Cyclic Pattern')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.legend()
plt.show()
```
